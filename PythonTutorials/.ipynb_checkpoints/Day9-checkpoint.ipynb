```python
# Soniya Answer
import requests
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
for idx,img in enumerate(jiphone['images']):
    response = requests.get(jiphone['images'][idx])
    img = Image.open(BytesIO(response.content))
    plt.subplot(3, 3, idx+1), plt.imshow(img, cmap='gray')
plt.show()
```

```python
# 1 1 1 3 3 3 2
name = "1 1 1 3 3 3 2"
demo = name.split()
#print(demo)
count = 0
for i in range(0, len(demo)-1):
    if demo[i] != demo[i+1] or demo[i] != demo[i-1]:
        count += 1
count
```

```python
# num = 3
# case1: 1 1 1 3 3 3 2 - 4
# case2: 1 3 1 1 1 - 3
# case3: 5 5 5 5 - 0
num = int(input())
for j in range(num):
    name = input()
    demo = name.split()
    #print(demo)
    count = 0
    for i in range(0, len(demo)-1):
        if demo[i] != demo[i+1] or demo[i] != demo[i-1]:
            count += 1
    print(count)
```
